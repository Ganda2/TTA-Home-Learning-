 Altair Visualisation

```python
import altair as alt
import pandas as pd
import numpy as np
```

```python
# load the dataset
holiday_dest = pd.read_csv('holiday_destinations.csv')
```

```python
holiday_dest.head()
```

```python
holiday_dest.describe()
```

```python
alt.Chart(holiday_dest).mark_point().encode(
    x='country', 
    y= 'destination'
).interactive()
```

```python
pointPlot = alt.Chart(holiday_dest).mark_point().encode(
    x='country', 
    y= 'destination',
    color = 'feedback'
)
pointPlot
```

```python
circlePlot = alt.Chart(holiday_dest).mark_circle().encode(
    x='country', 
    y='destination',
    size = 'star_rating',
    tooltip=['destination', 'star_rating'],
).interactive()
circlePlot
```

```python
barsplt = alt.Chart(holiday_dest).mark_bar().encode(
    y='country:N',
    #color='star_rating',
    x='count(country):Q'

)
barsplt 
```

```python
barPlot = alt.Chart(holiday_dest).mark_bar().encode(
    x='country', 
    y='avg_reviews_score',
)
barPlot

```

```python
stackbarPlot = alt.Chart(holiday_dest).mark_bar().encode(
    x='avg_reviews_score', 
    y='country',
    color='star_rating:N',
    tooltip=['destination', 'star_rating']

)
stackbarPlot
```

```python
circlePlot & stackbarPlot
```

```python
brush = alt.selection_interval()#
```

```python
ccirclePlot = alt.Chart(holiday_dest).mark_circle().encode(
    x='country', 
    y='avg_reviews_score',
    size = 'star_rating',
    tooltip=['destination', 'star_rating','country'],
    color=alt.condition(brush, 'star_rating:N', alt.value('lightgray')),
).add_selection(
    brush
)

circlePlot
```
